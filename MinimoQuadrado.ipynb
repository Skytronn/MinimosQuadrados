```python
#Carregamento do conjunto de dados
data = pd.read_excel('valores.xlsx')


ano= np.array(data['Ano'])
Y= np.array(data['Patrimonio(R$ bilhoes)'])
t= np.array(data['Tempo'])
tY = np.array(data['tY'])
t2 = np.array(data['t2'])
n= 11
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

Y = Y.reshape(-1,1)
t = t.reshape(-1,1)
ano = ano.reshape(-1,1)

linreg = LinearRegression().fit(t,Y)

b= linreg.coef_

print('Coeficiente angular da regressão linear (a): {}'
     .format(linreg.coef_))
```

```python
#Descobrindo 'a' que é o coeficiente linear da reta.
```

```python
linreg = LinearRegression().fit(t,Y)

a = linreg.intercept_

print("Coeficiente linear da regressão linear:")
print(a)
```

```python
figure = plt.figure(figsize=(9,6))
plt.scatter(t , Y, marker= 'o', s=50, alpha=0.8)
plt.plot( t , Y, 'k--', color='grey')
plt.plot(t , linreg.coef_ * t + linreg.intercept_, 'r-')
plt.title('Patrimônio líquido')
plt.xlabel('Tempo')
plt.ylabel('Patrimônio líquido(R$ bilhoes)')
plt.xticks(np.arange(min(t), max(t)+1, 1.0))
plt.xticks(t, [ 2007,2008,2009,2010,2011,2012,2013,2014,2015,2016,2017],rotation = 20 )
plt.show()
```
